```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values
x = np.linspace(0, 20, 100)
# Calculate the sine of each x value
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, '-b', label='Sine wave')
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
